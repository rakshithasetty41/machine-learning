```python
# Step 1: Import libraries
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,r2_score
import numpy as np
```

```python
# Step 2: Load dataset
data = {
    "Area": [1200, 1500, 1800, 2400, 3000, 3500, 4000, 4200, 4800, 5000],
    "Bedrooms": [2, 3, 3, 4, 4, 4, 5, 5, 5, 6],
    "Bathrooms": [1, 2, 2, 3, 3, 3, 4, 4, 5, 5],
    "Age":  [10, 8, 5, 3, 2, 1, 4, 5, 6, 2],
    "Price": [150000, 180000, 210000, 300000, 350000, 400000, 420000, 450000, 480000, 500000]
}
df=pd.DataFrame(data)
df
```

```python
# Step 3: Clean & explore
df.info()
```

```python
df.describe()
```

```python
#step 4:input and target variables
x=df[["Area","Bedrooms","Bathrooms","Age"]]
y = df["Price"]

```

```python
x.shape
```

```python
y.shape
```

```python
# Step 5: Split into train and test sets
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
y_train
```

```python
print(x_test.shape)
print(x_train.shape)
print(y_test.shape)
print(y_train.shape)
```

```python
# Step 6: Train Linear Regression model
model=LinearRegression()
model.fit(x_train,y_train)
# Step 7: Predict
y_predict=model.predict(x_test)
y_predict
```

```python
y_test
```

```python

# Step 8: Evaluate model
print("mse",np.sqrt(mean_squared_error(y_test,y_predict)))
print("r2",r2_score(y_test,y_predict))
```

```python
# Step 9: Interpret coefficients
comp=pd.DataFrame({
    "Actual Price": y_test.values,
    "Predicted Price": np.round(y_predict, 2)})
comp
```

```python
coef_df = pd.DataFrame({
    "Feature": x.columns,
    "Coefficient": model.coef_})
coef_df
```

```python
# Step 10: Save the model
import joblib
joblib.dump(model,'house_price_model.pkl') #dump is a write/save
print("Model saved successfully!")
```

```python
loaded_model=joblib.load('house_price_model.pkl')
print("Model loaded successfully!")
```
