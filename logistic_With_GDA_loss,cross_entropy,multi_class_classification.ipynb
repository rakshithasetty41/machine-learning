```python
inputsdf=df.iloc[:,:-2]
inputs=inputsdf.to_numpy()
inputs
```

```python
originals=inputs.copy()
```

```python
def scale(x):
  return (x-x.mean())/x.std()
```

```python
def scalematrix(x):
  for i in range(x.shape[1]):
    x[:,i]=scale(x[:,i])
  return x
```

```python
scaled_inputs=scalematrix(inputs)
scaled_inputs
```

```python
def scaletest(x):
  stds=[]
  for i in range(x.shape[1]):
    stds.append(x[:,i].std())
  return stds
```

```python
def prediction(x,w):
  return 1/(1+np.exp(-x.dot(w)))
```

```python
def mse(y, ycap):
    return ((y-ycap)**2).mean()
```

```python
def derivative(p):
  return p*(1-p)
```

```python
conv = 0.00000001
ploss = 0
flag =0
for i in range(100000):
   ycap = prediction(X, W)
   e = Y - ycap
   closs = mse(Y, ycap)
   if i%1000==0:
      print(f"current loss after {i+1} iterations {closs}")
   delta = derivative(ycap)
   W += X.T.dot(delta) * 0.002
   if np.abs(ploss-closs) < conv:
     print(f"Training completed after {i+1} iterations ")
     flag = 1
     break
   ploss = closs
if flag==0:
   print("Training not completed , run few more iterations ")
```

```python
# part2
# improve accuracy with GDA.
# lets Rebuilding.
np.random.seed(101)
W = np.random.random((X.shape[1], Y.shape[1])) * 2 -1
W
```

```python
ploss =0
flag =0
conv = 0.0000001
for i in range(100000):
    ycap = prediction(X, W)
    e = Y - ycap
    closs = mse(Y , ycap)
    if i%1000 == 0:
        print(f"current loss after {i+1} iterations {closs}")

    delta =  e * derivative(ycap)   # here is change when comepared with previous training loop.
    W += X.T.dot(delta) * 0.02
    if np.abs(ploss-closs) <= conv:
        print(f"Training Completed after {i+1} iterations ")
        flag=1
        break
    ploss = closs
if flag==0:
   print("Training not Completed run few more iterations ")
```

```python
def crossentropy(y,ycap):
  y=np.array(y)
  ycap=np.array(ycap)
  left=y*np.log(ycap)
  rgt=(1-y)*np.log(1-ycap)
  lrsum=left+rgt
  s=lrsum.sum()
  ce=-s/len(y)
  return ce
```

```python
# one hot transformation --> convert numerical labels into binary array.
#       0 1 2
#----------------
# 0 --> 1 0 0
# 1 --> 0 1 0
# 2 --> 0 0 1
def toBinary(nlabels):
   arr = []
   for  i in nlabels:
       barr = np.zeros(3)
       barr[i] = 1
       arr.append(barr)
   return np.array(arr)
```
