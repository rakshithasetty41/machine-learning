```python
import pandas as pd
data = {
    "Area": [1200, 1500, 1800, 2400, 3000, 3500, 4000, 4200, 4800, 5000],
    "Bedrooms": [2, 3, 3, 4, 4, 4, 5, 5, 5, 6],
    "Bathrooms": [1, 2, 2, 3, 3, 3, 4, 4, 5, 5],
    "Age":  [10, 8, 5, 3, 2, 1, 4, 5, 6, 2],
    "Price": [150000, 180000, 210000, 300000, 350000, 400000, 420000, 450000, 480000, 500000]
}

df = pd.DataFrame(data)
df
```

```python
# Save dataset to CSV and Excel
df.to_csv("house_data.csv",index=False)
df.to_excel("house_data.xlsx", index=False)
```
