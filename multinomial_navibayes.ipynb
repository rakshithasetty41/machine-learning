```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python

# sample dataset
data ={

       "text":[
           "free money offer now",
           "win lottery cash prize",
           "meeting schedule today",
           "project deadline tomorrow",
           "exclusive offer just for you",
           "lets discuss work progress"
       ],
       "label":["spam","spam","not spam","not spam","spam","not spam"]
}
```

```python
# CountVectorizer is a tool used to convert text data into numerical format so machine learning model can understand
```

```python
df=pd.DataFrame(data)
df
```

```python
from numpy import vectorize
vectorizer=CountVectorizer()
x=vectorizer.fit_transform(df['text'])
x
```

```python
y=df['label']
y
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.55, random_state=2)
nb=MultinomialNB()
nb.fit(x_train,y_train)
y_pred=nb.predict(x_test)
y_pred
```

```python
print("Accuracy:",accuracy_score(y_test,y_pred))
print("Confursion Matrix:",confusion_matrix(y_test,y_pred))
```

```python
custom_message = ["Congratulations, you've won a free vacation"]
print(custom_message)
custom_vector = vectorizer.transform(custom_message)
prediction = nb.predict(custom_vector)
print("Prediction for custom message:", "Spam" if prediction[0] == 1 else "Not Spam")
```
