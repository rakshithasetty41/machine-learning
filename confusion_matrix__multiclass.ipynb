```python
# confusion matrix for multi class.
```

```python
y = [0, 1,2,0,1,2,2,1,1,0,0,2,1,0]
len(y)
```

```python
ycap = [0,1,2,1,0,1,0,2,1,2,0,2,1,1]
len(ycap)
```

```python
import numpy as np
```

```python
cm=np.zeros((3,3))
cm
```

```python
for i ,j in zip(y,ycap):
  cm[i,j]+=1
cm
```

```python
def confusion(y,ycap):
  n=len(set(y))
  cm=np.zeros((n,n))
  for i,j in zip(y,ycap):
    cm[i,j]+=1
  return cm
```

```python
confusion(y,ycap)
```

```python
yt = [0, 1, 2, 3, 1, 2, 3, 0, 0]
ycapt = [0, 1, 2, 3, 1, 1, 2, 3, 1]
confusion(yt, ycapt)
```

```python
P1=cm[1,1]/cm[:,1].sum()
R1=cm[1,1]/cm[1].sum()
P2=cm[2,2]/cm[:,2].sum()
R2=cm[2,2]/cm[2].sum()
```

```python
P=[P0,P1,P2]
R=[R0,R1,R2]
list(zip(P,R))
```

```python
def metrics(cm):
  p=[]
  r=[]
  f1=[]
  for i in range(cm.shape[0]):
    pr=cm[i,i]/cm[:,i].sum()
    re=cm[i,i]/cm[i].sum()
    f=2/(1/pr+1/re)
    p.append(pr)
    r.append(re)
    f1.append(f)
  return np.array((p,r,f1)).T
```

```python
metrics(cm)
```
