```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
# 1. Define sample data (manually crafted points)

X = np.array([

[1, 2], [1.5, 1.8], [2, 2.2],

[8, 8], [8.5, 7.5], [9, 8.2],

[0.5, 9], [1.2, 8.5], [1.8, 9.5],

[7, 1], [6.5, 1.5], [7.5, 0.5]

])
```

```python
X
```

```python
X.shape
```

```python
#2. Visualize original points

plt.scatter(X[:, 0], X[:, 1], s=50)

plt.title("Manually Defined Data")

plt.grid(True)

plt.show()
```

```python
# 3. Apply KMeans clustering (choosing 4 clusters)

kmeans = KMeans (n_clusters=4, random_state=42)

# random_state=<any number> --> seed value.

kmeans.fit(X)

y_kmeans = kmeans.predict(X)
```

```python
y_kmeans
```

```python
#4. Plot clusters and centroids

plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, cmap='viridis', s=50)

centroids = kmeans.cluster_centers_

plt.scatter(centroids [:, 0], centroids [:, 1], c='red', s=200, marker='X', label='Centroids')

plt.title("K-Means Clustering Result")

plt.legend()

plt.grid(True)

plt.show()
```

```python
import numpy as np

import matplotlib.pyplot as plt

from scipy.cluster.hierarchy import dendrogram, linkage

from sklearn.cluster import AgglomerativeClustering

# Agglomarative is used for Heirarchical clustering.

# clusters with in clusters.



```

```python
# 1. Manually defined dataset

X = np.array([

[1, 2], [1.5, 1.8], [2, 2.2],
[8, 8], [8.5, 7.5], [9, 8.2],
 [0.5, 9], [1.2, 8.5], [1.8, 9.5],
 [7, 1], [6.5, 1.5], [7.5, 0.5]

])
```

```python
#2. Dendrogram using Ward linkage
linked = linkage(X, method='ward')

plt.figure(figsize=(10,5))

dendrogram (linked,
            orientation='top',
            distance_sort='descending',
            show_leaf_counts=True)



plt.title("Dendrogram (Ward's Method)")

plt.grid(True)

plt.show()

#3. Hierarchical clustering (Agglomerative)

cluster = AgglomerativeClustering(n_clusters=4, linkage='ward')

# No affinity parameter neede

labels=cluster.fit_predict(X)

#4. Visualize clustered output

plt.scatter (X[:, 0], X[:, 1], c=labels, cmap='viridis', s=100)

plt.title("Agglomerative Clustering Output")
plt.grid(True)
plt.show()
```
